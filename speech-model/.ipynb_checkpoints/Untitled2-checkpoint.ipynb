```python
def get_pair_index(data_list, match_query, position):
    for i, pair in enumerate(data_list):
        if pair[position] == match_query:
            return i
    return -1

def sec_to_frame(sec, fs):
    return int(float(sec)*float(fs))

def frame_to_ms(frame, fs):
    return int(float(frame)*1.0/float(fs))

max_sample_mult = 1.5
min_sample_mult = 0.5

max_noise = 1
min_noise = 0

min_pitch = -3
max_pitch = 3

min_rms = 0.3
max_rms = 0.7

x_train = []
y_train = []

i = 0

y_labels = alphabets[:]
y_labels.append('other')
y_labels.append('silent')
print(y_labels)

for _ in tqdm(range(100)):
    for file in labeled_audio:
        file_name = file[0]
        labels = file[1]

        signal, fs = sf.read(dataset_folder_path + file_name)

        if len(signal.shape) == 2:
            signal = np.array([x[0] for x in signal])

        for alphabet in alphabets:
            start, stop = labels[get_pair_index(labels, alphabet, 0)][1]
            start = sec_to_frame(start, fs)
            stop = sec_to_frame(stop, fs)
#             print(file_name, alphabet, stop, start)
            sample_frame_num = stop - start
    #         print(signal.shape)
            new_sample_frames = min(fs - 1000, random.randrange(int(sample_frame_num*min_sample_mult), int(sample_frame_num*max_sample_mult)))
            rate = (sample_frame_num)/new_sample_frames
#             print(rate)
            sample = signal[start:stop]
            sample = librosa.effects.time_stretch(sample, rate=rate)

#             start = int(start * rate)
#             stop = int(stop * rate)
            sample_frame_num = sample.shape[0]

            time_left = sec_to_frame(1, fs) - (sample_frame_num)
            if time_left < 0:
                print("error")
    #         

            pre_time = random.randrange(0, time_left)
            post_time = time_left - pre_time

            n_steps = random.randrange(0, 10000)*(max_pitch - min_pitch)/10000 + min_pitch
            sample = librosa.effects.pitch_shift(sample, sr=fs, n_steps=n_steps)

            alphabet_rms = np.average(librosa.feature.rms(y=sample))
#             print(alphabet_rms, file_name, start/fs, stop/fs)
            amp = random.randrange(0, 10000)*(max_rms - min_rms)/10000 + min_rms
            sample *= amp/alphabet_rms

            alphabet_rms = np.average(librosa.feature.rms(y=sample))
            STD_n= random.randrange(0, 10000)*alphabet_rms*0.03/10000
            noise=np.random.normal(0, STD_n, sec_to_frame(1, fs))
            noisy_sample = np.concatenate((noise[:pre_time], sample + noise[pre_time : pre_time + sample_frame_num], noise[pre_time + sample_frame_num:]))

            sample_16k = librosa.resample(noisy_sample, orig_sr=fs, target_sr=16000)
            new_fs = 16000

#             sample_16k /= np.max(sample_16k)

#             fig, ax = plt.subplots()
#             fig.supxlabel('Time')
#             fig.set_size_inches(15, 5)
#             ax.plot(np.linspace(0, sample_16k.shape[0]/fs, sample_16k.shape[0]), sample_16k)

            mfcc = speechpy.feature.mfcc(sample_16k, sampling_frequency=new_fs, frame_length=0.040, frame_stride=0.02,
                                 num_filters=40, fft_length=256, low_frequency=0, high_frequency=None)
            
            x_train.append(mfcc)
            
            empty_arr = np.zeros(12)
            empty_arr[y_labels.index(alphabet)] = 1
            y_train.append(empty_arr[:])
            
            if i%997:
                sf.write(f"./samples/{alphabet}_{start/fs}_{stop/fs}_{rate}_{n_steps}_{i}_{file_name}.wav", sample_16k, new_fs)
            i += 1
            
        STD_n= random.randrange(int(10000*0.3), int(10000*1))/10000
        noise=np.random.normal(0, STD_n, sec_to_frame(1, 16000))
        
        mfcc = speechpy.feature.mfcc(noise, sampling_frequency=new_fs, frame_length=0.040, frame_stride=0.02,
                                 num_filters=40, fft_length=256, low_frequency=0, high_frequency=None)
        
        x_train.append(mfcc)
        
        empty_arr = np.zeros(12)
        empty_arr[y_labels.index('other')] = 1
        y_train.append(empty_arr[:])
        
        if i%997:
            sf.write(f"./samples/other_{i}_{file_name}.wav", noise, 16000)
        i += 1
        
        STD_n= random.randrange(0, int(10000*0.03))/10000
        noise=np.random.normal(0, STD_n, sec_to_frame(1, 16000))
        
        mfcc = speechpy.feature.mfcc(noise, sampling_frequency=new_fs, frame_length=0.040, frame_stride=0.02,
                                 num_filters=40, fft_length=256, low_frequency=0, high_frequency=None)
        
        x_train.append(mfcc)
        
        empty_arr = np.zeros(12)
        empty_arr[y_labels.index('silent')] = 1
        y_train.append(empty_arr[:])
        
        if i%997:
            sf.write(f"./samples/none_{i}_{file_name}.wav", noise, 16000)
        i += 1
x_train = np.array(x_train)
y_train = np.array(y_train)
```
